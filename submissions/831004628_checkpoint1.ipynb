````python
#Data Pre-Processing, always assume these functions are done in-place (for efficiency reasons)

def remove_unnamed_cols(df):
    df.drop(columns=[c for c in ["Unnamed: 0"] if c in df.columns], inplace=True)


def remove_duplicates(df, key_cols=None):
    if key_cols is None:
        key_cols = ["EventId", "EventYear", "MatchId"]  # default tournament-scoped key
    key_cols = [c for c in key_cols if c in df.columns]

    # Only drop duplicates if we actually have key columns
    if key_cols:
        df.drop_duplicates(subset=key_cols, inplace=True, keep="first")

def coerce_dates(df):
    for c in ["StartDate", "EndDate", "PlayerTeam1.RankDate", "PlayerTeam2.RankDate"]:
        if c in df.columns:
            df[c] = pd.to_datetime(df[c], errors="coerce")

#percent columns are not necessary since precision is bad, 
#and we are provided with divisor and dividend, so we can compute this ourselves
def drop_percent_cols(df):
    percent_cols = [c for c in df.columns if c.endswith(".Percent")]
    df.drop(columns=percent_cols, inplace=True)

def preprocess_raw_tennis(raw_tennis_df):
    remove_unnamed_cols(raw_tennis_df)
    remove_duplicates(raw_tennis_df)
    coerce_dates(raw_tennis_df)
    drop_percent_cols(raw_tennis_df)
    return raw_tennis_df

cleaned_tennis_df = preprocess_raw_tennis(raw_tennis_df)
````

### Data Pre-Processing (In-Place Cleaning)

In this step, I cleaned the raw tennis dataset using a small preprocessing pipeline designed to be **efficient** and **repeatable**. All transformations are performed **in-place** (modifying the original DataFrame directly) to avoid unnecessary copies and reduce memory overhead.

The preprocessing includes:

- **Removing extra index columns:** Some CSV exports include a redundant `"Unnamed: 0"` column (usually from a saved index). I drop it if it exists.
- **Dropping duplicate matches:** I remove duplicate rows using a tournament-scoped identifier (`EventId`, `EventYear`, `MatchId`) to ensure each match appears only once in the dataset.
- **Converting date fields into proper datetime types:** Columns such as `StartDate`, `EndDate`, and each player's ranking date are coerced into `datetime` format, with invalid values converted to `NaT`.
- **Removing percent columns:** Columns ending in `.Percent` are dropped because percent values often have low precision, and the dataset already contains the numerator and denominator needed to recompute these percentages more accurately.

Finally, I wrapped all of these cleaning steps into a single function (`preprocess_raw_tennis`) and applied it to the raw dataset to produce `cleaned_tennis_df`.


# EDA 

````python
# Set-wise Match Statistics
df = cleaned_tennis_df
def safe_div(a, b):
    a = pd.to_numeric(a, errors="coerce")
    b = pd.to_numeric(b, errors="coerce")
    return np.where((b > 0) & (~pd.isna(b)), a / b, np.nan)

def extract_set_rate_and_total(df, team_prefix, set_i, base):
    d = f"{team_prefix}.Sets[{set_i}].{base}.Dividend"
    v = f"{team_prefix}.Sets[{set_i}].{base}.Divisor"

    if d not in df.columns or v not in df.columns:
        return None

    dividend = pd.to_numeric(df[d], errors="coerce")
    divisor  = pd.to_numeric(df[v], errors="coerce")

    out = pd.DataFrame({
        "set_number": set_i + 1,
        "team": team_prefix,
        "rate": np.where(divisor > 0, dividend / divisor, np.nan),
        "total": divisor,                      # <-- the "how many" count
        "has_total": divisor.notna(),
        "has_valid_rate": (divisor > 0) & dividend.notna(),
    })
    return out

def compute_stats_for_all_stats_columns_with_totals(tennis_df, min_rate_n=1, min_total_n=1):
    """
    For every stat base with Dividend/Divisor pairs, compute:
      - rate_mean/std and rate_n (where divisor > 0)
      - total_mean/std and total_n (where divisor is not NaN)

    Output columns:
      [stat_base, set_number,
       rate_mean, rate_std, rate_n,
       total_mean, total_std, total_n]
    """
    df_local = tennis_df

    pattern = re.compile(r"^PlayerTeam[12]\.Sets\[\d+\]\.(?P<base>.+)\.(Dividend|Divisor)$")

    bases = set()
    for c in df_local.columns:
        m = pattern.match(c)
        if m:
            bases.add(m.group("base"))
    bases = sorted(bases)

    rows = []
    for base in bases:
        pieces = []
        for team in ["PlayerTeam1", "PlayerTeam2"]:
            for set_i in range(6):
                chunk = extract_set_rate_and_total(df_local, team, set_i, base)
                if chunk is not None:
                    pieces.append(chunk)

        if not pieces:
            continue

        long = pd.concat(pieces, ignore_index=True)

        summary = (long.groupby("set_number")
                      .agg(
                          rate_mean=("rate", "mean"),
                          rate_std=("rate", "std"),
                          rate_n=("has_valid_rate", "sum"),   # counts rows with divisor>0

                          total_mean=("total", "mean"),
                          total_std=("total", "std"),
                          total_n=("has_total", "sum"),       # counts rows with divisor not NaN
                      )
                      .reset_index())

        # Optional coverage filters
        summary = summary[(summary["rate_n"] >= min_rate_n) & (summary["total_n"] >= min_total_n)].copy()
        if summary.empty:
            continue

        summary.insert(0, "stat_base", base)
        rows.append(summary)

    if not rows:
        return pd.DataFrame(columns=[
            "stat_base","set_number",
            "rate_mean","rate_std","rate_n",
            "total_mean","total_std","total_n"
        ])

    out = pd.concat(rows, ignore_index=True)

    # Sort by coverage
    out["total_rate_n_for_stat"] = out.groupby("stat_base")["rate_n"].transform("sum")
    out = out.sort_values(["total_rate_n_for_stat", "stat_base", "set_number"],
                          ascending=[False, True, True]).drop(columns=["total_rate_n_for_stat"])

    return out

results = compute_stats_for_all_stats_columns_with_totals(df, min_rate_n=1)
print(results.head(30))
````

### Set-Wise Match Statistics (Rate + Coverage Analysis)

In this section, I analyzed the dataset at the **set level** to understand how match statistics are recorded across different sets (especially to investigate whether stats are aggregated into set 1 or split across sets 1–3+).

Many of the tennis statistics are stored as **Dividend/Divisor pairs**, meaning the dataset provides both:
- a **numerator** (Dividend), and  
- a **denominator / opportunity count** (Divisor)

This allows me to compute a more accurate rate than relying on any precomputed percentage fields.

What the code does:

- **Identifies all available stat types automatically:**  
  Using a regex pattern, I scan the DataFrame for any columns matching the structure  
  `PlayerTeamX.Sets[i].<StatName>.Dividend` and `.Divisor`.  
  This extracts every stat “base” (e.g., first serve %, break points won, etc.) without hardcoding column names.

- **Extracts set-wise rates and totals for both teams:**  
  For each stat base, I loop through:
  - both teams (`PlayerTeam1` and `PlayerTeam2`)
  - up to 6 sets (to support long matches)

  For each set/team/stat, I compute:
  - `rate = Dividend / Divisor` (only when divisor > 0)
  - `total = Divisor` (interpreted as the number of opportunities / attempts)
  - flags indicating whether the stat is present and whether the computed rate is valid

- **Aggregates into a coverage table by set number:**  
  After converting everything into a long-format table, I group by `set_number` and compute summary statistics:
  - mean and standard deviation of the rate (`rate_mean`, `rate_std`)
  - how many rows had a valid rate (`rate_n`)
  - mean and standard deviation of the total (`total_mean`, `total_std`)
  - how many rows contained any divisor at all (`total_n`)

- **Sorts results by coverage:**  
  The final output is sorted so that the most commonly available stats appear first, making it easier to inspect which stats are well-recorded and how they vary across sets.

Finally, I print the first 30 rows of this results table to quickly examine which statistics exist per set and how complete the set-wise coverage is.


````python
# looking specifically at the average number of points played per set since the set-wise stastistics looked suspicious

def extract_set_total(df, team_prefix, set_i, base, key_cols):
    v = f"{team_prefix}.Sets[{set_i}].{base}.Divisor"
    if v not in df.columns:
        return None

    out = pd.DataFrame({
        "set_number": set_i + 1,
        "total": pd.to_numeric(df[v], errors="coerce"),
    })

    # Attach match keys row-wise (same length as df), so no index-mismatch later
    for k in key_cols:
        out[k] = df[k].values

    return out

# Total points played in a set (per player) = Divisor of TotalPointsWon
base = "Stats.PointStats.TotalPointsWon"
key_cols = [c for c in ["EventId", "EventYear", "MatchId"] if c in df.columns]

pieces = []
for team in ["PlayerTeam1"]:
    for set_i in range(6):
        chunk = extract_set_total(df, team, set_i, base, key_cols)
        if chunk is not None:
            pieces.append(chunk)

long = pd.concat(pieces, ignore_index=True)

# Sum P1 + P2 totals => points played per match per set
match_points = (
    long.groupby(key_cols + ["set_number"], dropna=False)["total"]
        .sum(min_count=1)
        .reset_index(name="points_played")
)

# Average points played by set number (pandas mean ignores NaNs)
set_summary = (
    match_points.groupby("set_number")["points_played"]
        .agg(avg_points_played="mean", n="count")
        .reset_index()
        .sort_values("set_number")
)

plt.figure()
plt.bar(set_summary["set_number"].astype(str), set_summary["avg_points_played"])
plt.xlabel("Set #")
plt.ylabel("Average points played (both players)")
plt.title("Average Points Played by Set Number")
plt.show()

set_summary
````

### Investigating Points Played Per Set (Sanity Check on Set-Wise Stats)

Because the set-wise statistics looked suspicious, I ran a targeted check on a quantity that should behave consistently across sets: **the number of points played per set**.

The dataset stores many “counts” as a **Divisor** field. For point totals, the stat base  
`Stats.PointStats.TotalPointsWon` provides a divisor that functions as the total number of point opportunities recorded for that team in that set. I use this as a proxy for *points played*.

What the code does:

- **Extracts per-set totals from the Divisor column:**  
  For each set index (0–5), I pull the column  
  `PlayerTeam1.Sets[i].Stats.PointStats.TotalPointsWon.Divisor`  
  (if it exists) and convert it to numeric. Each extracted chunk is labeled with:
  - `set_number` (1-indexed)
  - `total` (the per-team count for that set)
  - match identifiers (`EventId`, `EventYear`, `MatchId`) so rows can be grouped safely later

- **Builds a long-format table across sets:**  
  All per-set chunks are concatenated into one long DataFrame (`long`) so I can group and summarize cleanly.

- **Computes points played per match per set:**  
  I group by match keys + `set_number` and sum the totals to get a per-match estimate of  
  `points_played` for each set.

- **Summarizes average points played by set number:**  
  I then aggregate across all matches to compute:
  - `avg_points_played`: mean points played for each set number
  - `n`: number of match-set rows contributing to that mean

- **Visualizes the result:**  
  Finally, I create a bar chart showing the **average points played** for Set 1, Set 2, Set 3, etc.  
  Set 1 was abnormally large indicating that statistics are usually aggregated into set 1 when recorded.


````python
# -----------------------
# 1) Define “has set stats” indicator
# -----------------------
def add_has_set_stats_flags(df):
    # A “stat exists” proxy (pick a stat divisor that should exist if per-set stats exist)
    s1_p1 = "PlayerTeam1.Sets[1].Stats.PointStats.TotalPointsWon.Divisor"
    s1_p2 = "PlayerTeam2.Sets[1].Stats.PointStats.TotalPointsWon.Divisor"
    s2_p1 = "PlayerTeam1.Sets[2].Stats.PointStats.TotalPointsWon.Divisor"
    s2_p2 = "PlayerTeam2.Sets[2].Stats.PointStats.TotalPointsWon.Divisor"

    def col_notna(col):
        return df[col].notna() if col in df.columns else pd.Series(False, index=df.index)

    df = df.copy()
    df["has_set2_stats"] = col_notna(s1_p1) | col_notna(s1_p2)  # set index 1 => “2nd set”
    df["has_set3_stats"] = col_notna(s2_p1) | col_notna(s2_p2)  # set index 2 => “3rd set”
    df["has_set2_and_3_stats"] = df["has_set2_stats"] & df["has_set3_stats"]
    return df

df2 = add_has_set_stats_flags(df)

print("Rows:", len(df2))
print("Has set2 stats:", int(df2["has_set2_stats"].sum()))
print("Has set3 stats:", int(df2["has_set3_stats"].sum()))
print("Has set2+set3 stats:", int(df2["has_set2_and_3_stats"].sum()))
print("Pct with set2+3:", float(df2["has_set2_and_3_stats"].mean()))

# -----------------------
# 2) Helper: coverage tables by group
# -----------------------
def coverage_by_col(df, col, flag="has_set2_and_3_stats", top_n=25, min_count=50):
    """
    Returns a table showing, for each category in `col`,
    - total rows
    - flag_count (rows where flag is True)
    - flag_rate
    """
    if col not in df.columns:
        return None

    g = (df.groupby(col, dropna=False)[flag]
           .agg(total="count", flag_count="sum", flag_rate="mean")
           .reset_index()
           .sort_values(["flag_rate", "flag_count"], ascending=False))

    # optional: filter out tiny groups
    g = g[g["total"] >= min_count].head(top_n)
    return g

# -----------------------
# 3) Run comparisons on likely drivers
# -----------------------
drivers = [
    "EventYear",
    "EventType",
    "TournamentName",
    "Singles",
    "Doubles",
    "InOutdoor",
    "Court",
    "TourneyLocation",
    "Round.ShortName",
    "Round.LongName",
    "PlayerTeam1.PlayerCountryCode",
    "PlayerTeam2.PlayerCountryCode",
]

tables = {}
for c in drivers:
    t = coverage_by_col(df2, c, flag="has_set2_and_3_stats", top_n=30, min_count=50)
    if t is not None and len(t):
        tables[c] = t
        print("\n=== Coverage by", c, "===\n")
        print(t.to_string(index=False))
````

### Identifying When Per-Set Stats Exist (and What Predicts It)

In this section, I tried to determine **which matches actually contain per-set statistics** beyond set 1, and what variables appear to explain that behavior.

#### 1) Create “has set stats” flags
Because the dataset does not provide an explicit indicator for “this match has per-set stats,” I built a proxy flag based on whether a statistic that *should* exist for a set is present and non-null.

- I chose `Stats.PointStats.TotalPointsWon.Divisor` as the proxy because it’s a fundamental count-like field that should be recorded whenever per-set stats are recorded.
- I check this divisor for both teams in:
  - set index `[1]` (which corresponds to **Set 2**)
  - set index `[2]` (which corresponds to **Set 3**)

From these checks, I add three boolean columns:
- `has_set2_stats`: Set 2 stats exist for at least one team
- `has_set3_stats`: Set 3 stats exist for at least one team
- `has_set2_and_3_stats`: both Set 2 and Set 3 stats exist (stronger evidence of true per-set recording)

I then print counts and percentages to quantify how common per-set stats are overall.

#### 2) Build coverage tables by group
Next, I wrote a helper (`coverage_by_col`) that produces a small “coverage table” for any categorical column.  
For each category value (e.g., a given year or tournament), it computes:
- `total`: number of matches in that group
- `flag_count`: number of matches where `has_set2_and_3_stats` is True
- `flag_rate`: fraction of matches in that group with set 2+3 stats

To avoid noise from tiny groups, I filter out categories with fewer than `min_count` matches and keep only the top `top_n` categories with the highest coverage rates.

#### 3) Test likely “drivers” of set-wise availability
Finally, I run the coverage table across a list of columns that might plausibly explain why set 2/3 stats appear, including:

- **Time effects:** `EventYear`
- **Tournament structure:** `EventType`, `TournamentName`, `Round.ShortName`, `Round.LongName`
- **Match type:** `Singles`, `Doubles`
- **Surface/conditions:** `InOutdoor`, `Court`
- **Location:** `TourneyLocation`
- **Player metadata:** both players’ country codes

This produces side-by-side comparisons that help answer questions like:
- Is this mainly a **year / recording-era** change?
- Is it concentrated in specific **tournaments** or **rounds**?
- Does it depend on **match type** (singles vs doubles) or **court conditions**?

Overall, this step helps determine whether per-set stats availability is **systematic (provider/era/tournament-driven)** rather than random missingness.


# (E) Initial Insights and Direction

So one of the first insights I've sussed out is that the stats data for set[0] (meaning the first set) tends to have about triple the number of points than it should (compared to the other sets). We also see that there are only a couple thousand rows where sets 2 and 3 are actually defined, despite the fact that all tennis matches should have at least 3 sets.  
This implies that in any match where data for the second set is not available, the first set simply has all of the match statistics aggregated.  
In data where we do have 2nd and 3rd sets, it may be difficult to tell whether or not the first set contains aggregated data.  
To find out what might be causing these 2nd and 3rd sets to exist, I looked at what all of these matches had in common.  
Judging from the coverage tables generated, it would appear that this has to do with some kind of shift in how the data was recorded in 2025. In every year before 2025, there weren't any matches with recorded 2nd and 3rd match statistics, and in 2025 approximately 37% of matches contained data for 2nd and 3rd sets, and in 2026 48%.  
Since this is quite possibly not a random occurence / error, we need to still include these matches which contain 2nd and 3rd match statistics.  
Maybe some of this problem can be solved by using anomaly detection techniques? Although we have yet to cover this in class.    

Overall however, there seems to be a lot of intact data here (~94,000 matches) with well structured, tabular data. This yields itself well to large scale machine learning techniques, especially decision trees.
